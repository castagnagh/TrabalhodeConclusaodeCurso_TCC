```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Colunas de entrada
X_cols = [
    'dados_pessoais: sexo',
    'dados_pessoais: faixa_etaria', 'dados_pessoais: posicao_quadra',
    'experiencia: tempo_padel', 'experiencia: quantidade_treinos',
    'experiencia: quantidade_jogos', 'experiencia: participa_torneios',
    'experiencia: categoria_em_torneios', 'habilidade_tecnica: saque',
    'habilidade_tecnica: voleio', 'habilidade_tecnica: smash',
    'habilidade_tecnica: movimentacao', 'composicao_jogo: defesa_ataque',
    'composicao_jog0: controle_ritmo', 'objetivo: estilo_duplas',
    'objetivo: foco', 'composicao_jogo: saque',
    'composicao_jogo: usa_retoque', 'composicao_jogo: voleio_ataque',
    'composicao_jogo: voleio_defesa_contensao',
    'composicao_jogo: bandeja_retomada_rede', 'composicao_jogo: smash',
    'composicao_jogo: troca_bola_fundo_forehand',
    'composicao_jogo: troca_bola_fundo_backhand',
    'composicao_jogo: devolucao_saque',
    'composicao_jogo: saida_parede_defesa',
    'composicao_jogo: saida_parede_ataque', 'composicao_jogo: movimentacao',
    'composicao_jogo: posicionamento', 'dados_pessoais: autorizacao'
]

# Colunas de saída
y_cols = ['autoavaliacao: categoria']

X = df[X_cols]
y = df[y_cols]
```

```python
#Para KNN, get_dummies é mais seguro, porque KNN mede distância Euclidiana e números arbitrários
# podem distorcer a interpretação de “distância” entre categorias.

#Para árvores (Decision Tree, Random Forest), usar map ou LabelEncoder funciona bem, porque árvores
#não são sensíveis à escala numérica.

X = pd.get_dummies(X, columns=[
    'dados_pessoais: posicao_quadra',
    'composicao_jogo: defesa_ataque',
    'composicao_jog0: controle_ritmo',
    'objetivo: estilo_duplas',
    'objetivo: foco'
])

#colunas criadas após a execução do comando get_dummies

#'dados_pessoais: posicao_quadra_direita', 'dados_pessoais: posicao_quadra_esquerda',
#'dados_pessoais: posicao_quadra_não sei ainda', 'dados_pessoais: posicao_quadra_tanto faz',
#'composicao_jogo: defesa_ataque_ainda não sei', 'composicao_jogo: defesa_ataque_atacar',
#'composicao_jogo: defesa_ataque_defender', 'composicao_jog0: controle_ritmo_não',
#'composicao_jog0: controle_ritmo_não sei responder', 'composicao_jog0: controle_ritmo_sim',
#'objetivo: estilo_duplas_fixas', 'objetivo: estilo_duplas_não sei', 'objetivo: estilo_duplas_rotativas',
#'objetivo: foco_as três alternativas', 'objetivo: foco_as três primeiras opções em conjunto',
#'objetivo: foco_competir', 'objetivo: foco_melhorar como jogador',
#'objetivo: foco_melhorar como jogador e socializar',
#'objetivo: foco_melhorar na parte técnica, avançado em categoria e competição',
#'objetivo: foco_não sei', 'objetivo: foco_praticar uma atividade fisica e socializar',
#'objetivo: foco_socializar',
#'objetivo: foco_socializar, fazer atividade fisica e melhorar como jogador',
#'objetivo: foco_todas alternativas', 'objetivo: foco_todas as alternativas',
#'objetivo: foco_todos', 'objetivo: foco_todos as opcoes acima', 'objetivo: foco_um pouco de todos']
```

```python
print(X.columns.tolist())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 3. Separar treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 4. Treinar KNN
knn = KNeighborsClassifier(n_neighbors=5)  # você pode testar outros valores de k
knn.fit(X_train, y_train)
```

```python
# 5. Fazer previsões
y_pred = knn.predict(X_test)
```

```python
# 6. Avaliar
print("Acurácia:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
